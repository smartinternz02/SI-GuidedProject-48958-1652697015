```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Line Graph
x = np.linspace(-4, 4, 100)
y = 2*x + 5
plt.plot(x, y, '-r')
plt.show()
```

```python
#Scatter Plot
x = np.random.randint(-5, 50, 15)
y = np.random.randint(-5, 100, 15)
plt.scatter(x, y)
plt.show()
```

```python
#Bar Graph
data_dict = {'Math':20, 'Physics':32, 'Biology':55, 'Chemistry': 18}
courses = list(data_dict.keys())
values = list(data_dict.values())
fig = plt.figure(figsize = (10, 5))
#  Bar plot
plt.bar(courses, values, color ='blue',
        width = 0.5)
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Marks Distribuition")
plt.show()
```

```python
#Histogram
val = np.random.normal(500, 50, 80)

plt.hist(val)
plt.show() 
```

```python
#Count Plot
df = sns.load_dataset('tips')
sns.countplot(x ='sex', data = df)
plt.show()
```

```python
#Heatmap
data = pd.DataFrame(np.random.random((8, 8)), columns=["a","b","c","d","e","f","g","h"])
p1 = sns.heatmap(data)
```

```python
#Box Plot
np.random.seed(10)
data = np.random.normal(150, 40, 150)

fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```

```python
#Pair Plot
data2 = sns.load_dataset('tips')
sns.pairplot(data2, hue ='day')
plt.show()
```

```python
#Joint Plot
data3 = sns.load_dataset("exercise")
sns.jointplot(x = "id", y = "pulse", kind = "kde", data = data3)
plt.show()
```

```python
#Kernel Density Plot
data4 = sns.load_dataset('tips')
data4.tip.plot.kde(color='red')
plt.title('KDE-Density plot for Tip')
plt.show()
```
